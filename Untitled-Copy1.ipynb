```python
import pandas as pd
import numpy as np
```

```python
def str_to_float(x):
    try:
        float(x)
        return x
    except:
        return float(x.replace('$','').replace(',', ''))
```

```python
degrees_that_pay_back_file = r"degrees-that-pay-back (1).csv"
salaries_by_college_type_file = r"salaries-by-college-type (1).csv"
df_degrees_that_pay_back = pd.read_csv(degrees_that_pay_back_file)
df_salaries_by_college_type = pd.read_csv(salaries_by_college_type_file)
```

```python
salary_col = ['Starting Median Salary', 'Mid-Career Median Salary', 'Mid-Career 10th Percentile Salary', 'Mid-Career 25th Percentile Salary', 'Mid-Career 75th Percentile Salary', 'Mid-Career 90th Percentile Salary']
```

```python
df_degrees_that_pay_back[salary_col] = df_degrees_that_pay_back[salary_col].apply(np.vectorize(str_to_float))
```

```python
df_degrees_that_pay_back.head()
```

```python
df_degrees_that_pay_back.describe()
```

```python
#create a list for index of max number in salary column 
max_list = df_degrees_that_pay_back[salary_col].idxmax(axis = 0).to_list()
#index of max number in salary column 
min_list = df_degrees_that_pay_back[salary_col].idxmin(axis = 0).to_list()
```

```python
#create a datafram
summary = pd.DataFrame()
#colnum
summary[salary_col] = ''
#row
summary['index'] = ['max','max_value','min', 'min_value']
#set the first colnum as index
summary.index = summary['index']
```

```python
#this for loop aims to find the maximum and minimum value in each salary colnum and label them using discrete value of major
for index, col in enumerate(salary_col):
    summary.loc['max', col] = df_degrees_that_pay_back.loc[max_list[index],'Undergraduate Major']
    summary.loc['max_value', col] = df_degrees_that_pay_back.loc[max_list[index], col]
    summary.loc['min', col] = df_degrees_that_pay_back.loc[min_list[index],'Undergraduate Major']
    summary.loc['min_value', col] = df_degrees_that_pay_back.loc[min_list[index], col]
summary = summary.drop(["index"], axis = 1)
```

```python
# Clean the format. Convert currency format to float format.
df_salaries_by_college_type[salary_col] = df_salaries_by_college_type[salary_col].apply(np.vectorize(str_to_float))
```

```python
no_nan_col = ['School Name', 'School Type', 'Starting Median Salary', 'Mid-Career Median Salary', 'Mid-Career 25th Percentile Salary', 'Mid-Career 75th Percentile Salary']
```

```python
max_list_by_school = df_salaries_by_college_type[salary_col].idxmax(axis = 0).to_list()
min_list_by_school = df_salaries_by_college_type[salary_col].idxmin(axis = 0).to_list()
```

```python
summary_by_school = pd.DataFrame()
summary_by_school[salary_col] = ''
summary_by_school['index'] = ['max','max_value','min', 'min_value']
summary_by_school.index = summary_by_school['index']
```

```python
for index, col in enumerate(salary_col):
    summary_by_school.loc['max', col] = df_salaries_by_college_type.loc[max_list_by_school[index],'School Name']
    summary_by_school.loc['max_value', col] = df_salaries_by_college_type.loc[max_list_by_school[index], col]
    summary_by_school.loc['min', col] = df_salaries_by_college_type.loc[min_list_by_school[index],'School Name']
    summary_by_school.loc['min_value', col] = df_salaries_by_college_type.loc[min_list_by_school[index], col]
summary_by_school = summary_by_school.drop(["index"], axis = 1)
```
